```python
import os
import numpy as np
import pandas as pd
```

```python
# Create a files variable that contains all of our data files.
files = os.listdir("data")
```

```python
files[:5]
```

```python
# Read in data from 2000
year_2000 = pd.read_csv("./data/annual_aqi_by_county_2000.csv")
year_2000.head()
```

```python
# function to read data
def process_data(file):
    df = pd.read_csv("./data/" + file)
    return df
```

```python
file_list = [process_data(file) for file in files]
```

```python
file_merge = pd.concat(file_list, axis=0)
```

```python
file_merge
```

```python
file_merge.info()
```

```python
# check null value
file_merge.isna().sum()
```

```python
file_merge.to_csv('data/air_quality.csv', index=False)
```
